```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv("../data/tech_reviews.csv")
print(df.shape)
df.head()
```

```python
# Distribution des labels
print(df["label"].value_counts())
print(f"\nSplit train/test : {df['split'].value_counts().to_dict()}")
```

```python
df["label"].value_counts().plot(kind="bar", color=["red", "green"])
plt.title("Distribution Positif / Négatif")
plt.xticks([0, 1], ["Négatif", "Positif"], rotation=0)
plt.ylabel("Nombre d'avis")
plt.show()
```

```python
df["text_length"] = df["text"].apply(lambda x: len(x.split()))
print(df.groupby("label")["text_length"].describe())

df.groupby("label")["text_length"].hist(alpha=0.7, bins=20)
plt.title("Longueur des avis par classe")
plt.xlabel("Nombre de mots")
plt.legend(["Négatif", "Positif"])
plt.show()
```
